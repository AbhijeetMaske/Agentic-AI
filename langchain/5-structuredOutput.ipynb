### Structured Output
Model can be requested to provide their response in format matching given schema. This is useful for ensuring the output can be easily parsed and used in subsequent processing. LangChain supports multiple schema types and methods for enforcing strcutured output.

### Pydantic 
Pydantic models provide the richest feature set with field validation, description and nested structures.

```python
from langchain.chat_models import init_chat_model
os.environ["openrouter_api_key"] = openrouter_api_key
model = init_chat_model(
    "openrouter:deepseek/deepseek-v4-flash",)
model
```

```python
from pydantic import BaseModel, Field

class movie(BaseModel):
    title: str = Field(..., description="The title of the movie")
    year: int = Field(..., description="The year the movie was released")
    genre: str = Field(..., description="The genre of the movie")
    director: str = Field(..., description="The director of the movie")
    rating: float = Field(..., description="The rating of the movie on a scale from 0 to 10")

```

```python
model_with_structured_output = model.with_structured_output(movie)
model_with_structured_output
```

```python
model.invoke("Please provide information about the movie Inception.")
```

```python
model_with_structured_output.invoke("Please provide information about the movie Pirates of the Caribbean.")
```

### Message output alongside parsed structure

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    """a movie with a title, year, genre, director, and rating"""
    title: str = Field(..., description="The title of the movie")
    year: int = Field(..., description="The year the movie was released")
    genre: str = Field(..., description="The genre of the movie")
    director: str = Field(..., description="The director of the movie")
    rating: float = Field(..., description="The rating of the movie on a scale from 0 to 10")

model_with_structured_output = model.with_structured_output(movie,include_raw=True)
model_with_structured_output.invoke("dhoom")
```

### Nested Structure



```python
from pydantic import BaseModel, Field

class Actor(BaseModel):
  name: str
  role: str

class MovieDetails(BaseModel):
  title : str
  year: int
  cast: list[Actor]
  genres: list[str]
  budget: float | None = Field (None, description="budget in millions USD")

model_with_structured_output= model.with_structured_output(MovieDetails)
model_with_structured_output.invoke("provide the details about inception")
```

### TypedDict

TypedDict provides a simpler alternative using python's build-in typing, Ideal when you don't need runtime validation

```python
from typing_extensions import TypedDict, Annotated

class MovieDict(TypedDict):
  """A movie with details"""
  title: Annotated [str, ...,"the title of movie"]
  year: Annotated[int,...,"The year of movie release"]
  director: Annotated[str,...,"The director of movie"]
  rating: Annotated[int,...,"The movie's rating out of 10"]

model_with_typeDict = model.with_structured_output(MovieDict)
response = model_with_typeDict.invoke("Provide the details of movie avengers")
response
```

```python
from pydantic import BaseModel, Field

class Actor(TypedDict):
  name: str
  role: str

class MovieDetails(TypedDict):
  title : str
  year: int
  cast: list[Actor]
  genres: list[str]
  budget: float | None = Field (None, description="budget in millions USD")

model_with_structured_output= model.with_structured_output(MovieDetails)
model_with_structured_output.invoke("provide the details about avengers")
```

```python
from langchain.chat_models import init_chat_model
os.environ["openrouter_api_key"] = openrouter_api_key
model = init_chat_model("openrouter:deepseek/deepseek-v4-flash",)
```

```python
from pydantic import BaseModel, Field
from langchain.agents import create_agent

class ContactInfo(BaseModel):
  """Contact information for a person."""
  name: str = Field (description="The name of the person")
  email: str = Field(description="The email ID of the person")
  phone : int = Field(description="the phone number of the person")

agent = create_agent(
  model = init_chat_model("openrouter:deepseek/deepseek-v4-flash",),
  response_format = ContactInfo #Autoselects providerStrategy
)

result = agent.invoke({
  "messages": [{"role":"user","content":"Extract contact info from: john doe, john@example.com, 1234567890"}]
})
result
#print(result["Structured_response"])
#ContactInfo(name='john doe',email='john@example.com', phone='1234567890' )
```

```python
### Typedict
from pydantic import BaseModel, Field
from langchain.agents import create_agent

class ContactInfo(TypedDict):
  """Contact information for a person."""
  name: str = Field (description="The name of the person")
  email: str = Field(description="The email ID of the person")
  phone : int = Field(description="the phone number of the person")

agent = create_agent(
  model = init_chat_model("openrouter:deepseek/deepseek-v4-flash",),
  response_format = ContactInfo #Autoselects providerStrategy
)

result = agent.invoke({
  "messages": [{"role":"user","content":"Extract contact info from: john doe, john@example.com, 1234567890"}]
})
```

```python
### data Class

from dataclasses import dataclass
from langchain.agents import create_agent

@dataclass
class ContactInfo:
  """Contact information for a person."""
  name: str # The name of the person
  email: str # The email ID of the person 
  phone : int # The phone number of the person

agent = create_agent(
  model = init_chat_model("openrouter:deepseek/deepseek-v4-flash",),
  response_format = ContactInfo #Autoselects providerStrategy
)

result = agent.invoke({
  "messages": [{"role":"user","content":"Extract contact info from: john doe, john@example.com, 1234567890"}]
})

result["structured_response"]
```
