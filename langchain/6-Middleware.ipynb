### Middleware
Middleware provide a way to more tighlty control waht happens inside the agent. Middleware is useful for the following:
-  Tracking agent behavior with logging, analytics and dubugging
-  Transforming prompts, tool selection, and output formatting.
-  Adding retries, fallback and early termination logic
-  apply rate limit, guardrails and PII detection.
